```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
data=sklearn.datasets.load_boston()
```

```python
data
```

```python
df =pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
df['price']=data.target
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
correlation=df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap="Purples")
```

```python
x=df.drop(['price'],axis=1)
y=df['price']
```

```python
x

```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_train)
```

```python
prediction
```

```python
score_1=metrics.r2_score(y_train,prediction)
score_2=metrics.mean_absolute_error(y_train,prediction)
```

```python
score_1
```

```python
score_2
```

```python
test_prediction=model.predict(x_test)
```

```python
score_1=metrics.r2_score(y_test,test_prediction)
score_2=metrics.mean_absolute_error(y_test,test_prediction)
```
